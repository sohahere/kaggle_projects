```python
import pandas as pd
import re
import string
import emoji
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# --- Start with a fresh DataFrame ---
df = pd.read_csv('/content/IMDB Dataset.csv', on_bad_lines='skip', engine='python')

# 1. Drop duplicate rows
df.drop_duplicates(inplace=True)

# 2. Convert all text to lowercase
df['review'] = df['review'].str.lower()

# 3. Remove HTML tags
def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
df['review'] = df['review'].apply(remove_html_tags)

# 4. Normalize chat words
chat_words_map = {
    'lol': 'laughing out loud',
    'brb': 'be right back',
    'btw': 'by the way',
    'omg': 'oh my god',
    'idk': 'i don\'t know',
    'u': 'you',
    'r': 'are',
    'y': 'why'
}
def normalize_chat_words(text):
    sorted_chat_words = sorted(chat_words_map.keys(), key=len, reverse=True)
    pattern = re.compile(r'\b(' + '|'.join(re.escape(word) for word in sorted_chat_words) + r')\b')
    def replace_word(match):
        return chat_words_map[match.group(0)]
    return pattern.sub(replace_word, text)
df['review'] = df['review'].apply(normalize_chat_words)

# 5. Remove punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))
df['review'] = df['review'].apply(remove_punctuation)

# 6. Remove Stop Words
nltk.download('stopwords', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True) # Ensure this is also downloaded
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    word_tokens = word_tokenize(text)
    filtered_sentence = [w for w in word_tokens if not w in stop_words]
    return " ".join(filtered_sentence)

print("Applying stop word removal... This might take a moment.")
df['review'] = df['review'].apply(remove_stopwords)

# 7. Convert emoji to text
print("Applying emoji to text conversion using the 'emoji' library...")
df['review'] = df['review'].apply(emoji.demojize)

print("Review after ALL preprocessing steps (first review):")
print(df['review'][0])
```

## Logistic Regression with GridSearchCV

### Subtask:
Train a Logistic Regression model using GridSearchCV to find the optimal hyperparameters. Evaluate its performance on the test set and display key metrics.


**Reasoning**:
The subtask requires training a Logistic Regression model with GridSearchCV, evaluating its performance, and displaying key metrics. This involves importing specific libraries, defining a hyperparameter grid, initializing the model, running GridSearchCV, and then printing the best parameters, classification report, and accuracy score, which can all be done in a single code block.



```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score

# 2. Define a dictionary param_grid_lr containing hyperparameters to tune
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga']
}

# 3. Initialize a LogisticRegression model
# Increased max_iter for potential convergence issues, especially with 'saga'
logreg = LogisticRegression(max_iter=1000, random_state=42)

# 4. Instantiate GridSearchCV
grid_search_lr = GridSearchCV(estimator=logreg,
                              param_grid=param_grid_lr,
                              cv=5,
                              scoring='accuracy',
                              n_jobs=-1,  # Use all available processors
                              verbose=1)

print("Starting GridSearchCV for Logistic Regression...")

# 5. Fit the GridSearchCV object to your training data
grid_search_lr.fit(X_train, y_train)

print("GridSearchCV completed.")

# 6. Print the best parameters found by the grid search
print("\nBest parameters found:", grid_search_lr.best_params_)

# 7. Get the best estimator from the grid search
best_logreg_model = grid_search_lr.best_estimator_

# 8. Make predictions on the test set
y_pred_lr = best_logreg_model.predict(X_test)

# 9. Print the classification report
print("\nClassification Report for Best Logistic Regression Model:\n")
print(classification_report(y_test, y_pred_lr))

# 10. Print the accuracy score
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print(f"Accuracy Score for Best Logistic Regression Model: {accuracy_lr:.4f}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# 2. Initialize a CountVectorizer object
# Not specifying binary=True will count occurrences
# Limiting max_features for performance and readability
vectorizer_bow = CountVectorizer(max_features=5000)

# 3. Fit and transform the review data into a Bag of Words matrix
X_bow = vectorizer_bow.fit_transform(df['review'])

# 4. Print the shape of the resulting X_bow matrix
print("Shape of Bag of Words matrix:", X_bow.shape)

# 5. Retrieve the vocabulary (feature names)
vocabulary_bow = vectorizer_bow.get_feature_names_out()
print("Number of unique words (vocabulary size) for BoW:", len(vocabulary_bow))

# 6. Sum the columns of the X_bow matrix to get total count for each word
word_counts = X_bow.sum(axis=0)

# Create a DataFrame or Series from these counts
word_freq_df = pd.DataFrame({'word': vocabulary_bow, 'count': word_counts.tolist()[0]})

# 7. Display the total number of unique words (vocabulary size)
print(f"\nTotal number of unique words (vocabulary) found by BoW: {len(vocabulary_bow)}")

# 8. Display the top N most frequent words and their counts
N = 20 # You can change N to display more or fewer top words
print(f"\nTop {N} most frequent words and their counts:")
print(word_freq_df.sort_values(by='count', ascending=False).head(N).to_string(index=False))
```
